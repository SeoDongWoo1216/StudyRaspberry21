```python
m = folium.Map(location=[35.1175, 129.0903], zoom_start=12)
folium.Marker(location=[35.1175, 129.0903], popup='부경대학교 용당캠퍼스').add_to(m)
folium.Marker(location=[35.1781, 129.199141], popup='송정해수욕장',
            icon=folium.Icon(color='red', icon='remove-circle')).add_to(m)
m
```

```python
import json

m = folium.Map(location=[36.634445, 127.795277], zoom_start=7)  # 한국지도 전체표시
gfile = open('./data/province-simple.json', mode='r', encoding='utf-8').read()
jsondata = json.loads(gfile)
folium.GeoJson(jsondata, name="province").add_to(m)
folium.Marker(location=[36.1175, 127.795277], popup='지도중심').add_to(m)
m
```
